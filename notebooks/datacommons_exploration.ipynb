```python
import datacommons as dc
import datacommons_pandas as dc_pandas
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd
import dotenv
import requests

d_vals = dotenv.dotenv_values("../.env")
dc.set_api_key(d_vals["DC_API_KEY"])
```

```python
usa = "country/USA"


def get_variables(entity_dcid, api_key):
    url = f"http://api.datacommons.org/v1/variables/{entity_dcid}"
    params = {"key": api_key}

    response = requests.get(url, params=params)
    response.raise_for_status()  # Raise an error if the request fails
    return response.json()


print("get states")
states = dc.get_places_in([usa], "State")[usa]
vars = get_variables(states[0], d_vals["DC_API_KEY"])
```
